```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df = pd.get_dummies(data=data,columns=['purpose'],drop_first=True)
df.head()
```

```python
scaler.fit(df.drop(['credit.policy','not.fully.paid','purpose_credit_card','purpose_debt_consolidation','purpose_educational','purpose_home_improvement','purpose_major_purchase','purpose_small_business'], axis=1))
```

```python
scaled_data= scaler.transform(df.drop(['credit.policy','not.fully.paid','purpose_credit_card','purpose_debt_consolidation','purpose_educational','purpose_home_improvement','purpose_major_purchase','purpose_small_business'],axis=1))
```

```python
df2 = pd.DataFrame(scaled_data,columns=df.columns[1:12])
df2.head()
```

```python
fdata= pd.concat([df2,df[['credit.policy','not.fully.paid','purpose_credit_card','purpose_debt_consolidation','purpose_educational','purpose_home_improvement','purpose_major_purchase','purpose_small_business']]],axis=1)
fdata.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X = fdata.drop('not.fully.paid',axis=1)
y =fdata['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
prediction = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, prediction))
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
#Applying the Random Forest Model
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
